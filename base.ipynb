```python
#======== Python 3.8.10 =====================

import cv2
import mediapipe
import time
from HTModule import handDetector
import math
import numpy as np
import autopy
from ctypes import cast, POINTER
from comtypes import CLSCTX_ALL
from pycaw.pycaw import AudioUtilities, IAudioEndpointVolume
import pyautogui
import pyttsx3
from pyKey import pressKey, releaseKey, press, sendSequence, showKeys
import pydirectinput


cam_width, cam_height = 683, 384
scr_width, scr_height = autopy.screen.size()
#print(scr_width, scr_height)
cap = cv2.VideoCapture(0)
cap.set(3, cam_width)
cap.set(4, cam_height)

detector = handDetector(maxHands=2)

if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

ptime = 0


def fingers_up(positions):
    fingers = []
    for i in range(6):
        if positions[4*i-2][2] > positions[4*i][2]:
            fingers.append(1)
        else:
            fingers.append(0)
    
    return fingers


def text_to_speech(text, rate=150):
    engine = pyttsx3.init()
    engine.setProperty('rate', rate)  # Adjust the speed of speech
    engine.say(text)
    engine.runAndWait()



def DirStick(frame, positions1,  positions2):
    
    point_finger = positions2[8]
    middle_finger = positions2[12]
    x_value, y_value = positions2[8][1:]
    x = np.interp(x_value, (0, cam_width), (0, scr_width)) 
    y = np.interp(y_value, (0, cam_height), (0, scr_height))
    length = math.hypot((point_finger[1] - middle_finger[1]) + (point_finger[2] - middle_finger[2]))
    print(fingers_up(positions2))

    if fingers_up(positions1)[2] == 1 and fingers_up(positions1)[3] == 1 and fingers_up(positions1)[4] == 1 and fingers_up(positions1)[5] == 1:
        # cv2.circle(frame, positions1[0][1:], 10, (0,255,0), cv2.FILLED)
        cv2.text(frame, "FORWARD", (positions1[0][1], positions1[0][2]), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 3)
        releaseKey('DOWN')
        pressKey('UP')

    if fingers_up(positions1)[2] == 0 and fingers_up(positions1)[3] == 0 and fingers_up(positions1)[4] == 0 and fingers_up(positions1)[5] == 0:
        # cv2.circle(frame, positions1[0][1:], 10, (0,255,0), cv2.FILLED)
        cv2.text(frame, "REVERSE", (positions1[0][1], positions1[0][2]), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 3)
        releaseKey('UP')
        pressKey('DOWN')

    if int(positions2[7][1]) < int(positions2[0][1]):
        # cv2.circle(frame, positions2[8][1:], 10, (0,255,0), cv2.FILLED)

        # Press the 'RIGHT' key
        releaseKey('LEFT')
        pressKey('RIGHT')
            
        
        

    elif int(positions2[20][1]) > int(positions2[0][1]):
        # cv2.circle(frame, positions2[16][1:], 10, (0,255,0), cv2.FILLED)
        # Press the 'LEFT' key
        releaseKey('RIGHT')
        pressKey('LEFT')
        
        

    


def GearControl(frame, positions1):
    
    # if fingers_up(positions1)[2] == 0 and fingers_up(positions1)[3] == 0 and fingers_up(positions1)[4] == 0 and fingers_up(positions1)[5] == 0:
    #     cv2.circle(frame, positions1[0][1:], 10, (0,255,0), cv2.FILLED)
    #     pressKey('space')

    if positions1[7][1] < positions1[0][1]:
        # cv2.circle(frame, positions1[8][1:], 10, (0,255,0), cv2.FILLED)
        pressKey('DOWN')
        

    elif positions1[20][1] > positions1[0][1]:
        # cv2.circle(frame, positions1[16][1:], 10, (0,255,0), cv2.FILLED)
        pressKey('a')


while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Could not read frame.")
        break
    
    frame = detector.findHands(frame)
    NoHands = detector.findNoOfHands(frame)
    if NoHands == 2:
        positions_list = detector.findPosition(frame, handNo=0, draw=False)
        positions_list_1 = detector.findPosition(frame, handNo=1,  draw=False)
        if len(positions_list) != 0 :
            # print(positions_list_1[8][1])
            # print(positions_list_1[0][1])
            DirStick(frame, positions_list, positions_list_1)
            # GearControl(frame, positions_list)
    elif NoHands == 1:
        text_to_speech("Only One hand detected", 150)
        
    
    ctime = time.time()
    fps = 1/(ctime-ptime)
    ptime = ctime

    # cv2.putText(frame, f'FPS: {int(fps)}', (20,40), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 3)
    cv2.imshow("Webcam", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
